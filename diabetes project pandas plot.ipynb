```python
# 21 axvline to display the vertical line in min/max point of column 

Glucose_min=df["Glucose"].min()
Glucose_max=df["Glucose"].max()
plt.figure(figsize=(7,7))
df['Glucose'].hist(bins=10)
plt.title("Glucose-min distribution")
plt.axvline(Glucose_min,color="red")
plt.axvline(Glucose_max,color="yellow")
plt.show()
```

```python
# axhline to display the horizontal lines in mean/median of column
Glucose_mean=df["Glucose"].mean()
Glucose_median=df["Glucose"].median()
plt.figure(figsize=(7,7))
df['Glucose'].hist(bins=10)
plt.title("Glucose-min distribution")
plt.axhline(Glucose_mean,color="red")
plt.axhline(Glucose_median,color="yellow")
plt.show()
```
